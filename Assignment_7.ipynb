#### Part I

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Roll no. as random seed
r_no = 102317112
np.random.seed(r_no)

# Generate sales data (12 months,4 categories)
sales_data = np.random.randint(1000,5001,size = (12,4))

# Convert to DataFrame
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
categories = ["Electronics","Clothing","Home & Kitchen","Sports"]
df = pd.DataFrame(sales_data,columns = categories,index = months)

df
```

#### Part II

```python
# Display first 5 rows
print(df.head())
print()

# Summary statistics
print(df.describe())
print()

# Total sales per category
total_sales_category = df.sum()
print(total_sales_category)
print()

# Total sales per month
df["Total Sales"] = df.sum(axis=1)
df
print()

# Calculate average sales growth
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
df
print()

# Apply discount based on roll number
if r_no % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.90  # 10% discount
else:
    df["Clothing"] = df["Clothing"] * 0.85  # 15% discount

df
print()
```

#### Part III

```python
# Line plot for sales trends
plt.figure(figsize = (10,6))
for category in categories:
    plt.plot(df.index,df[category],marker = "o",label = category)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.show()

# Box plot for sales distribution
plt.figure(figsize = (8,6))
sns.boxplot(data = df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

### Q2.

```python
array = np.array([[1,-2,3],[-4,5,-6]])

# i. Element-wise absolute value
abs_values = np.abs(array)
print("Absolute Values:\n",abs_values)
print()

# ii. Percentiles
percentiles = np.percentile(array,[25,50,75],axis = None)  # Flattened array
percentiles_columns = np.percentile(array,[25,50,75],axis = 0)  # Per column
percentiles_rows = np.percentile(array,[25,50,75],axis = 1)  # Per row

print("Percentiles (Flattened) :",percentiles)
print("Percentiles (Per Column) :",percentiles_columns)
print("Percentiles (Per Row) :",percentiles_rows)
print()

# iii. Mean, Median, Standard Deviation
mean_all = np.mean(array)
median_all = np.median(array)
std_all = np.std(array)

mean_column = np.mean(array,axis = 0)
median_column = np.median(array,axis = 0)
std_column = np.std(array,axis = 0)

mean_row = np.mean(array,axis = 1)
median_row = np.median(array,axis = 1)
std_row = np.std(array,axis = 1)

print("Mean (All) :",mean_all)
print("Median (All) :",median_all)
print("Standard Deviation (All) :",std_all)
print()

print("Mean (Column) :",mean_column)
print("Median (Column) :",median_column)
print("Standard Deviation (Column) :",std_column)
print()

print("Mean (Row) :",mean_row)
print("Median (Row) :",median_row)
print("Standard Deviation (Row) :",std_row)
print()
```

### Q4.

```python
def swap_list_elements(list1,pos1,pos2):
    temp = list1[pos1]
    list1[pos1] = list1[pos2]
    list1[pos2] = temp
    return list1

list1 = [1,2,3,4,5]
print("Before Swap :",list1)
print("After Swap :",swap_list_elements(list1,1,3))  # Swapping elements at index 1 and 3
```

### Q5.

```python
def swap_elements_in_set(input_set,elem1,elem2):
    # Check if both elements exist in the set
    if elem1 not in input_set or elem2 not in input_set:
        return None

    # Convert set to list
    temp_list = list(input_set)

    # Get indices of elements
    index1 = temp_list.index(elem1)
    index2 = temp_list.index(elem2)

    # Swap elements
    temp_list[index1],temp_list[index2] = temp_list[index2],temp_list[index1]

    # Convert back to set and return
    return set(temp_list)

# Example usage
my_set = {1,2,3,4,5}
result = swap_elements_in_set(my_set,2,4)
print(f"Original set : {my_set}")
print(f"Set after swapping 2 and 4 : {result}")
```
