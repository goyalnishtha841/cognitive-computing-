```python
#question 4
import numpy as np
import matplotlib.pyplot as plt
arr=np.linspace(-10,10,100)
square_arr=arr*arr
plt.plot(arr,square_arr)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of x^2")
plt.legend()
plt.grid(True)
plt.show()

sined=np.sin(arr)
plt.plot(arr,sined,color="green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of sin x")
plt.legend()
plt.grid(True)
plt.show()

exp_values=np.exp(arr)
plt.figure(figsize=(8,4))
plt.plot(arr,exp_values,label="e^x", color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot of e^x")
plt.legend()
plt.grid(True)
plt.show()

log_values=np.log(arr+1)
plt.figure(figsize=(8,4))
plt.plot(arr,log_values,label="log(|x|+1)",color="black")
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot of log(|x|+1")
plt.legend()
plt.grid(True)
plt.show()





```

```python
first_initial="N"
last_initial="G"
X=ord(first_initial)+ord(last_initial)
sales=np.array([X,X+50,X+100, X+150, X+200])
print("Sales data:", sales)

tax_rate = ((X % 5) + 5) /100
tax_applied = sales * tax_rate 
sales_after_tax = sales - tax_applied
print('sales after tax',sales_after_tax)

discount_rates = np.where(sales < X+100, 0.05, 0.10)  # 5% if sales < X+100, else 10%
discount_applied = sales_after_tax * discount_rates
sales_after_discount = sales_after_tax - discount_applied
print('sales after discount',sales_after_discount)

weeks=3
weekly_sales = np.vstack([sales_after_discount] * weeks)
increase_factors = np.array([1.00, 1.02, 1.04]).reshape(-1, 1) 
weekly_sales = weekly_sales * increase_factors 
print("Weekly Sales with 2% Increase per Week:\n", weekly_sales)





```
